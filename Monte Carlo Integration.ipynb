```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Estimating integral from 0 to 1 of x^(5/2)dx using Classical Monte-Carlo Method
# Estimate integral = (1/N)*(sum from i=1 to N of U^(5/2)) where U ~ Unif(0,1)

num_of_simulations=500
m = []

for i in range(0, num_of_simulations):
    U1 = np.random.uniform(0,1,1000)
    mn = (1/len(U1))*np.sum([np.power(u1, 5/2) for u1 in U1])
    m.append(mn)
    
plt.grid()
plt.hist(m)
plt.show()
print('m = {}'.format(np.mean(m)))
print('The error between the analytical and the numerical is {}'.format(abs(np.mean(m)-(2/7))))
```

```python
# Estimating the same integral by using the geometrical approach
# Geom. Estimate = (1/N)*(sum from i=1 to N of Zi) where Zi = c, if U' <= h(U), and 0 otherwise
# c bounds h(x) in the interval (0,1), therefore, c=1
# U', U ~ Unif(0,1)

U1 = np.random.uniform(0,1,1000)
U2 = np.random.uniform(0,1,1000)

m2 = []

for i in range(0, num_of_simulations):
    U1 = np.random.uniform(0,1,1000)
    U2 = np.random.uniform(0,1,1000)
    mn2 = (1/len(U1))*np.sum([1 if u2 <= np.power(u1, 5/2) else 0 for u1, u2 in zip(U1,U2)])
    m2.append(mn2)
    
plt.hist(m2)
plt.grid()
plt.show()

print('m = {}'.format(np.mean(m2)))
print('The error between the analytical and the numerical is {}'.format(abs(np.mean(m2)-(2/7))))
```
